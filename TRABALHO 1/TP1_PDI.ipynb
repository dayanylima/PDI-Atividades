#### Geração de um checkerboard

```python
# Criando padrões preto (0's) e branco (1's)
B = np.zeros((10,10)) #  representar os 0's
W = np.ones((10,10))  #  representar os 1's

# Concatenando os padrões para formar uma linha
BW = np.concatenate((B,W), axis=1) 
WB = np.concatenate((W,B), axis=1)

# Repetindo a linha para formar um padrão completo
BW_PATTERN = np.tile(BW,5) 
WB_PATTERN = np.tile(WB,5) 

checkerboardLine = np.concatenate((BW_PATTERN, WB_PATTERN), axis=0) 
checkerboard = np.tile(checkerboardLine,(5,1)) 

checkerboardConvert = np.array(checkerboard, dtype='uint8') 
checkerboardPixel255 = checkerboardConvert * 255

plt.imshow(checkerboardPixel255, cmap='gray', vmin=0, vmax=255) 
plt.axis('on')
plt.show()
```

```python
# Criando padrões preto (0's) e branco (1's)
B = np.zeros((1,1)) #  representar os 0's
W = np.ones((1,1))  #  representar os 1's

# Concatenando as casas horizontalmente
houses = np.hstack((B, W))

plt.imshow(houses, cmap='gray', vmin=0, vmax=1)
plt.axis('on')  
plt.show()
```

```python
# Padrão BW (preto e branco)
B = np.zeros((10, 10))  # padrão preto
W = np.ones((10, 10))   # padrão branco
BW = np.concatenate((B, W), axis=1)

# Repetir o padrão BW horizontalmente 5 vezes
BW_tiled = np.tile(BW, (1, 5))

# Exibindo o padrão de repetição horizontal
plt.imshow(BW_tiled, cmap='gray', vmin=0, vmax=1)
plt.axis('on')  # Desativando os eixos
plt.show()
```

```python
# Criando padrões preto (0's) e branco (1's)
B = np.zeros((10,10)) #  representar os 0's
W = np.ones((10,10))  #  representar os 1's

# Concatenando os padrões para formar uma linha
BW = np.concatenate((B,W), axis=1) 
WB = np.concatenate((W,B), axis=1)

# Repetindo a linha para formar um padrão completo
BW_PATTERN = np.tile(BW,5) 
WB_PATTERN = np.tile(WB,5) 

checkerboardLine = np.concatenate((BW_PATTERN, WB_PATTERN), axis=0) 
checkerboard = np.tile(checkerboardLine,(5,1)) 

checkerboard[checkerboard == 1] = 127
checkerboardConvert = np.array(checkerboard, dtype='uint8') 


plt.imshow(checkerboardConvert, cmap='gray', vmin=0, vmax=255) 
plt.axis('on')
plt.show()
```
